# ***HISTOGRAMA DE IDADE POR GENERO***

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = '/content/healthcare_dataset.csv'
data = pd.read_csv(file_path)

bins = np.arange(0, 101, 10)

plt.figure(figsize=(10, 6))

colors = {'Male': 'blue', 'Female': 'pink'}

bar_width = 4.5

for i, gender in enumerate(data['Gender'].unique()):
    subset = data[data['Gender'] == gender]
    counts, _ = np.histogram(subset['Age'], bins=bins)
    plt.bar(bins[:-1] + i * bar_width, counts, width=bar_width, color=colors[gender], label=gender, align='edge', edgecolor='black')

plt.title('Histograma de Idade por Gênero')
plt.xlabel('Idade (de 10 em 10 anos)')
plt.ylabel('Frequência')
plt.legend(title='Gênero')
plt.grid(axis='y')

plt.xticks(bins)
plt.show()
```

# ***GRÁFICO DE PARES CARACTERISTICAS COMUNS DO CANCER DE MAMA***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

csv_file_path = '/content/breast-cancer.csv'
df = pd.read_csv(csv_file_path)
df_filtered = df[['radius_mean', 'texture_mean', 'concavity_mean', 'diagnosis']].copy()
df_filtered['diagnosis'] = df_filtered['diagnosis'].astype('category')
sns.set(style="ticks")
pairplot = sns.pairplot(df_filtered, hue='diagnosis', diag_kind='kde', markers=["o", "s"])
plt.suptitle('Gráfico de características comuns e câncer de mama', y=1.02)
plt.show()
```

# ***DISPERSÃO COM REGRESSÃO SOBRE A CORRELAÇÃO DE TAMANHO E PREÇO***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

df = pd.read_csv('kc_house_data.csv')

fig, ax = plt.subplots(1, 2)

sns.regplot(df, x=df['sqft_living'], y=df['price'], order=1 ,scatter_kws={"color": "black"}, ax=ax[0])
ax[0].set_title('Preço x Área da casa (Pés)')
ax[0].yaxis.set_major_formatter(ticker.EngFormatter())
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[0].axes.set_xlabel('')
ax[0].axes.set_ylabel('Preço')

sns.regplot(df, x=df['sqft_above'], y=df['price'], order=1
                    ,scatter_kws={"color": "black"}, ax=ax[1])
ax[1].set_title('Preço x Área da casa (Pés) s/ Porão')
ax[1].yaxis.set_major_formatter(ticker.EngFormatter())
ax[1].spines['top'].set_visible(False)
ax[1].spines['left'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].figure.set_size_inches((10, 5))
ax[1].axes.get_yaxis().set_visible(False)
ax[1].axes.set_xlabel('')

plt.figtext(0.191, 0.85, f"Correlação: {df.select_dtypes(include='number').corr()['price']['sqft_living']:.4}", fontdict={'color':'#505050'})
plt.figtext(0.578, 0.85, f"Correlação: {df.select_dtypes(include='number').corr()['price']['sqft_above']:.4}", fontdict={'color':'#505050'})

fig.text(0.52, 0.02, 'Tamanho', ha='center')
plt.plot()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Car_sales.csv')
df.tail(10)
```

# ***MONTADORAS QUE MAIS APARECEM***

```python
df['Manufacturer'].value_counts() \
    .head(10) \
    .plot(kind='bar', color='midnightblue')

for i, j in enumerate(df['Manufacturer'].value_counts().head(10)):
    plt.annotate(str(j), xy=(i, j), ha='center', va='bottom')

plt.grid(False)
plt.ylabel("Quantidade de vezes")
plt.xlabel("Montadora")
plt.title("10 montadoras que mais apareceram na lista",fontsize=17)
```

# ***CARROS DA VOLKSWAGEN COM MENOS VENDAS***

```python
df[df['Manufacturer'] == 'Volkswagen'].groupby('Model')['Sales_in_thousands'].sum() \
  .nsmallest(5) \
  .plot(kind='bar', color='firebrick')

for i, j in enumerate(df[df['Manufacturer'] == 'Volkswagen'].groupby('Model')['Sales_in_thousands'].sum().nsmallest(10)):
    plt.annotate(str(j), xy=(i, j), ha='center', va='bottom')

plt.grid(True)
plt.ylabel('Vendas (em milhares)')
plt.xlabel('Modelo')
plt.title('Os 5 modelos da Volkswagen com menos vendas', fontsize=17)
```
